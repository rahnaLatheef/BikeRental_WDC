# Data Analysis and Visualization with Python


## 1. Objective: Explore Washington D.C.Bike Rental Dataset

The dataset is downloaded from Kaggle (bike_train.csv). It provides hourly bike rental numbers in Washington D.C. for the years 2011 and 2012. The objective is to explore the effect that different weather and temporal factors have on the number of bikes rented.

## 2. Data Description

datetime - hourly date + timestamp

season - 1 = spring, 2 = summer, 3 = fall, 4 = winter

holiday - whether the day is considered a holiday

workingday - whether the day is neither a weekend nor holiday

weather - 1: Clear, Few clouds, Partly cloudy, Partly cloudy - 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist - 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds - 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog
                
temp - temperature in Celsius

atemp - "feels like" temperature in Celsius

humidity - relative humidity

windspeed - wind speed

casual - number of non-registered user rentals initiated

registered - number of registered user rentals initiated

count - number of total rentals

## Part I: Data Manipulation and Analysis

**1. Import the dataset into a pandas dataframe. Make sure that the date column is in pandas date time format.**

```python
import pandas as pd

# Load the dataset
data=pd.read_csv("bike_train.csv")
data
```

```python
# Convert the date column to date format
data["datetime"] = pd.to_datetime(data["datetime"])
data
```

**2. Check the data type of each column. How many rows are there in the dataset ? Does the dataset contain any missing values ?**

```python
data_types = data.dtypes
data_types
```

```python
total_rows = len(data)
total_rows
```

```python
missing_values = data.isnull().sum()
missing_values
```

The dataset does not contains any missing values

**3. Using the date column, create new columns for: year, month, day of the week and hour of the day.**

```python
data['year'] = data['datetime'].dt.year
data['month'] = data['datetime'].dt.month
data['day'] = data['datetime'].dt.dayofweek
data['hour'] = data['datetime'].dt.hour
data
```

**4. Rename the values in the season column to spring, summer, fall and winter**

```python

data['season'] = data['season'].replace({1: 'spring', 2: 'summer', 3: 'fall', 4: 'winter'})
data
```

**5. Calculate the total number of casual and registered bikes rented in the years 2011 and 2012.**

```python
# Filter data for the years 2011 and 2012
filtered_data = data[data['year'].isin([2011, 2012])]
filtered_data
```

```python
# Calculate total rentals for each category
total_registered_rentals = filtered_data.groupby('year')['registered'].sum()
total_casual_rentals = filtered_data.groupby('year')['casual'].sum()

print("Total Registered Rentals:")
print(total_registered_rentals)

print("\nTotal Casual Rentals:")
print(total_casual_rentals)
```

**6. Calculate the mean of the hourly total rentals count by season. Which season has the highest mean ?**

```python
# Group by 'season' and calculate the mean of the 'count' column
mean_by_season = data.groupby('season')['count'].mean()

# Identify the season with the highest mean
season_with_highest_mean = mean_by_season.idxmax()

print("Mean Rentals by Season:")
print(mean_by_season)

print("\nSeason with the Highest Mean:")
print(season_with_highest_mean)
```

**7. Are more bikes rented by registered users on working or non-working days ? Does the answer differ for non-registered users ? Is the answer the same for both years ?**

```python

# Create a new column 'user_type_category' based on existing 'casual' and 'registered' columns
data['user_type_category'] = data[['casual', 'registered']].idxmax(axis=1)

# Create a new column 'day_category' based on existing 'workingday' column
data['day_category'] = data['workingday'].map({0: 'non-working', 1: 'working'})

# Group by 'user_type_category' and 'day_category' and count the number of rentals
rental_counts = data.groupby(['user_type_category', 'day_category']).size().unstack()

# Print the result
print(rental_counts)
```

More bikes are rented by registered users on working days and the answer is differ for the non-registered users(Casual).

```python
# Filter the dataset to include only registered users.
registered_data = data[data['user_type_category'] == 'registered']

# Group the filtered data by the 'year' and 'day_category' columns
rentals_by_year_day = registered_data.groupby(['year', 'day_category']).size().unstack()

print(rentals_by_year_day)

```

```python
# Filter the dataset to include only registered users.
registered_data = data[data['user_type_category'] == 'casual']
# Group the filtered data by the 'year' and 'day_category' columns
rentals_by_year_day = registered_data.groupby(['year', 'day_category']).size().unstack()

print(rentals_by_year_day)
```

In both years also more bikes are rented by registered users on working days

**8. Which months in the year 2011 have the highest and the lowest total number of bikes rented ? Repeat for the year 2012.**

```python

# Filter data for the year 2011
bikes_2011 = data[data['year'] == 2011]

# Group by month and calculate the total number of bikes rented
monthly_counts_2011 = bikes_2011.groupby('month')['count'].sum()

# Find the month with the highest and lowest total bikes rented in 2011
highest_month_2011 = monthly_counts_2011.idxmax()
lowest_month_2011 = monthly_counts_2011.idxmin()

# Repeat the process for the year 2012

bikes_2012 = data[data['year'] == 2012]
monthly_counts_2012 = bikes_2012.groupby('month')['count'].sum()
highest_month_2012 = monthly_counts_2012.idxmax()
lowest_month_2012 = monthly_counts_2012.idxmin()

print(f"In 2011, the month with the highest total bikes rented is {highest_month_2011} "
      f"\n     and the month with the lowest total bikes rented is {lowest_month_2011}")

print(f"In 2012, the month with the highest total bikes rented is {highest_month_2012} "
      f"\n    and the month with the lowest total bikes rented is {lowest_month_2012}")

```

**9. Which type of weather have the highest and lowest mean of the hourly total rentals count ?**

```python

hourly_total_rentals = data.groupby('hour')['count'].sum()

# Merge the calculated totals back to the original DataFrame
data = data.merge(hourly_total_rentals, how='left', on='hour', suffixes=('', '_total'))


# Rename the new column
data.rename(columns={'rentals_count_total': 'hourly_total_rentals'}, inplace=True)
data
```

```python

#Group by weather type and calculate the mean of hourly total rentals:
weather_mean_rentals = data.groupby('weather')['count_total'].mean()

#Find the weather type with the highest and lowest mean rentals:
highest_weather = weather_mean_rentals.idxmax()
lowest_weather = weather_mean_rentals.idxmin()
print(f"The weather type with the highest mean hourly total rentals is {highest_weather}.")
print(f"The weather type with the lowest mean hourly total rentals is {lowest_weather}.")

```

**10. Calculate the correlation between the hourly total rentals count and all the numerical columns in the dataset. Which column has the highest correlation with the total rentals count ?**

```python

# Select only numerical columns
numerical_columns = data.select_dtypes(include=['number']) 

# Compute correlation matrix
correlation_matrix = numerical_columns.corr()  


# Extracting correlations with the 'count_total' column
correlations_with_rentals = correlation_matrix['count_total']

# Finding the column with the highest correlation
highest_correlation_column = correlations_with_rentals.abs().idxmax()

# Printing the correlations and the column with the highest correlation
print(correlations_with_rentals)
print(f"\nThe column with the highest correlation: {highest_correlation_column}")
```

**11. Create a new categorical column called day_period, which can take four possible values: night, morning, afternoon and evening. These values correspond to the following binning of the hour column: 0-6: night, 6-12: morning, 12-6: afternoon, 6-24:evening.**

```python

bins = [0, 6, 12, 18, 24]
labels = ['night', 'morning', 'afternoon', 'evening']

data['day_period'] = pd.cut(data['hour'], bins=bins, labels=labels, include_lowest=True, right=False)
data
```

**12. Generate a pivot table for the mean of the hourly total rentals count, with the index set to the day period and the column set to the working day column. What can you observe from the table ?**

```python

pivot_table = pd.pivot_table(data, values='count_total', index='day_period', columns='workingday', aggfunc='mean')
pivot_table
```

Observations from the pivot table:
    
Night Rentals:
Rentals during the night are slightly higher on working days (11394.77) compared to non-working days (11280.38).

Morning Rentals:
Morning rentals show identical values for both working and non-working days (95526.67).

Afternoon Rentals:
Afternoon rentals also exhibit identical values for both working and non-working days (136585.33).

Evening Rentals:
Evening rentals are consistent, with the same values for both working and non-working days (104204.50).

These observations suggest that there is a consistency in rental patterns across different day periods, with some marginal variations in night rentals during working and non-working days.

**3. Plot the the mean of the hourly total rentals count for working and non-working
days.**

```python
# Use groupby to calculate the mean for each category
mean_rentals = data.groupby('workingday')['count_total'].mean()

# Plot the bar chart
mean_rentals.plot(kind='bar', color=['blue', 'orange'])
plt.title('Mean Hourly Total Rentals for Working and Non-Working Days')
plt.xlabel('Working Day')
plt.ylabel('Mean Hourly Total Rentals')
plt.xticks([0, 1], ['Non-Working Day', 'Working Day'], rotation=0)
plt.show()
```

**4. Plot the the mean of the hourly total rentals count for the different months for
both years combined.**

```python
# Use groupby to calculate the mean for each month
mean_rentals_by_month = data.groupby('month')['count_total'].mean()

# Plot the bar chart
mean_rentals_by_month.plot(kind='bar', color='green')
plt.title('Mean Hourly Total Rentals Count by Month')
plt.xlabel('Month')
plt.ylabel('Mean Hourly Total Rentals Count')
plt.show()
```

**5. Plot the the mean of the hourly total rentals count for the different months for
both years separately in a multi-panel figure.**

```python
import seaborn as sns

# Create a multi-panel figure using seaborn's FacetGrid
g = sns.FacetGrid(data, col='year', height=5, aspect=1)

# Map a bar plot for mean hourly total rentals count for each month
g.map(sns.barplot, 'month', 'count_total', errorbar=None  )

# Set titles and labels
g.set_axis_labels('Month', 'Mean Hourly Total Rentals Count')
g.set_titles(col_template='{col_name} Year')

# Show the plot
plt.show()
```

The clusters of variables with high positive or negative correlations may have strong relationships.
